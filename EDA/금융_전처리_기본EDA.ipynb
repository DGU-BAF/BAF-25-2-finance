### 구간 설정
초단기 고객   
0~90일 (3개월 미만)   
초기 onboarding 실패, 단기 해지 패턴 확인   

단기 고객   
90~365일 (3개월~1년)   
가입 첫 해 안에서 이탈하는 고객   

중단기 고객   
365~1000일 (1~3년)   
KDE 분포에서 churn율이 아직 높은 구간   

----안정화 시작-------------------   
중기 고객
1000~1825일 (3~5년)   

중장기 고객   
1825~3650일 (5~10년)   
충성도 높은 집단   

장기 고객   
3650~6290일 (10년 이상)   
오래된 고객, churn 낮을 가능성   

관측 종료 고객   
6291일

```python
# 구간 정의
bins = [0, 90, 365, 1000, 1825, 3650, 6290, 7000]  
labels = ['<3m', '3-12m', '1-3yr', '3-5yr', '5-10yr', '10+yr', 'max_tenure']

df['tenure_group'] = pd.cut(
    df['days_tenure'], 
    bins=bins, 
    labels=labels, 
    right=False, 
    include_lowest=True
)

# 구간별 churn 통계 (count, mean)
tenure_stats = df.groupby('tenure_group')['churn'].agg(['count','mean']).round(4)
print(tenure_stats)

# -------------------------
# 📊 1. 구간별 churn율 + 고객 수 라벨
# -------------------------
ax = tenure_stats['mean'].plot(
    kind='bar', 
    figsize=(7,5), 
    title='Churn Rate by Refined Tenure Group'
)

ax.set_ylabel("해지율 비율")

# count 값을 bar 위에 표시
for i, (idx, row) in enumerate(tenure_stats.iterrows()):
    ax.text(i, row['mean'] + 0.005, f"{row['count']:,}", 
            ha='center', va='bottom', fontsize=9)

plt.show()

# -------------------------
# 📊 2. 구간별 유지/이탈 고객 수 (0 vs 1)
# -------------------------
tenure_counts = df.groupby(['tenure_group','churn']).size().unstack(fill_value=0)

tenure_counts.plot(
    kind='bar',
    stacked=False,   # 나란히 배치
    figsize=(7,5),
    color=['steelblue','orange'],
    title='Churn Count by Refined Tenure Group'
)

plt.ylabel('Number of Customers')
plt.xlabel('Tenure Group')
plt.legend(title='Churn', labels=['0 (유지)','1 (이탈)'])
plt.show()
```

churn 차이: 정상범위 11.51% vs 범위밖 7.99%

```python
### 가입 당시 나이(age_at_signup) 변수 생성

df['date_of_birth']  = pd.to_datetime(df['date_of_birth'],  errors='coerce')

# 1) 가입 당시 나이 계산 (년 단위, floor)
df['age_at_signup'] = ((df['cust_orig_date'] - df['date_of_birth']).dt.days // 365).astype('Int64')

# (선택) 기본 건강검진: 비현실 연령 플래그 & 클리핑(EDA용)
df['age_signup_outlier'] = (~df['age_at_signup'].between(18, 100)).astype('Int64')
df['age_at_signup_clean'] = df['age_at_signup'].clip(lower=18, upper=100)

print("age_at_signup 요약")
print(df['age_at_signup'].describe())
```

```python
# 2) 가입 당시 나이 분포 (절대 count 기준 히스토그램)
plt.figure(figsize=(9,4))
sns.histplot(df['age_at_signup_clean'].dropna(), bins=40, kde=False)
plt.title('Age at Signup — Distribution (count)')
plt.xlabel('Age at signup'); plt.ylabel('Number of customers')
plt.show()
```

```python
# 3) 연령대 그룹별 고객 수 & 이탈률
bins   = [18, 25, 35, 45, 55, 65, 75, 90, 101]  # 90~100을 한 구간으로 묶음
labels = ['18-24','25-34','35-44','45-54','55-64','65-74','75-89','90-100']
df['signup_age_group'] = pd.cut(df['age_at_signup_clean'], bins=bins, labels=labels, right=False)

ageg = df.groupby('signup_age_group')['churn'].agg(['count','mean']).round(4)
print("\n연령대별 고객 수 & churn율")
print(ageg)
```

```python
# (a) 연령대별 churn율 막대 + count 라벨
ax = ageg['mean'].plot(kind='bar', figsize=(9,4), title='Churn Rate by Age-at-Signup Group')
ax.set_ylabel('Churn rate')
for i, (idx, row) in enumerate(ageg.iterrows()):
    ax.text(i, row['mean'] + 0.005, f"{row['count']:,}", ha='center', va='bottom', fontsize=9)
plt.show()

# (b) 연령대별 유지/이탈 고객 수(절대치) 나란히
age_counts = df.groupby(['signup_age_group','churn']).size().unstack(fill_value=0)
age_counts.plot(kind='bar', stacked=False, figsize=(9,4),
                title='Customer Count by Age-at-Signup Group (0=유지, 1=이탈)')
plt.ylabel('Number of customers'); plt.xlabel('Age-at-signup group')
plt.show()

# 4) 가입연도별 가입 당시 나이 분포(코호트 보기)
df['cust_orig_year'] = df['cust_orig_date'].dt.year

plt.figure(figsize=(11,5))
sns.boxplot(data=df, x='cust_orig_year', y='age_at_signup_clean', showfliers=False)
plt.title('Age at Signup by Cohort Year (boxplot, no outliers)')
plt.xlabel('Cohort year'); plt.ylabel('Age at signup')
plt.xticks(rotation=45)
plt.show()

# (선택) 가입연도별 중앙값 라인으로 좀 더 간단히
med = df.groupby('cust_orig_year')['age_at_signup_clean'].median()
plt.figure(figsize=(10,4))
med.plot(marker='o')
plt.title('Median Age at Signup by Cohort Year')
plt.xlabel('Cohort year'); plt.ylabel('Median age at signup')
plt.show()
```

```python
# 소득 구간화
bins = [0, 50000, 100000, 200000, 300000]
labels = ['Low', 'Mid', 'High', 'Very High']
df['income_group'] = pd.cut(df['income'], bins=bins, labels=labels, right=False)

df['income_group'].value_counts()

# 그룹별 해지율 계산
churn_rate = df.groupby('income_group')['churn'].mean()

print(churn_rate)

plt.figure(figsize=(6,4))
sns.barplot(x=churn_rate.index, y=churn_rate.values)
plt.title("Churn Rate by Income Group")
plt.ylabel("Churn Rate")
plt.xlabel("Income Group")
plt.show()
```

```python
import scipy.stats as stats

contingency = pd.crosstab(df['income_group'], df['churn'])
chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("Chi-square Test")
print("Chi2:", chi2, "p-value:", p)
```
